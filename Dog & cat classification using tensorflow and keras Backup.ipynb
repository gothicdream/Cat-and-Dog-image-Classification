```python
import requests

HELIUS_API_KEY = "5482d58a-ae6f-4e45-8fb3-37caa1b118a7"

BASE_URL = f"https://mainnet.helius-rpc.com/?api-key={HELIUS_API_KEY}"

def get_recent_signatures(address, limit=5):
    payload = {
        "jsonrpc": "2.0",
        "id": 1,
        "method": "getSignaturesForAddress",
        "params": [address, {"limit": limit}]
    }
    res = requests.post(BASE_URL, json=payload)
    return [tx["signature"] for tx in res.json()["result"]]

# Example: Jupiter address
JUPITER_PROGRAM_ID = "JUP4Fb2cqiRUcaTHdrPC8h2gNsA2ETXiPDD33WcGuJB"
signatures = get_recent_signatures(JUPITER_PROGRAM_ID)
```

```python
def get_transaction_details(signature):
    payload = {
        "jsonrpc": "2.0",
        "id": 1,
        "method": "getTransaction",
        "params": [signature, {"encoding": "jsonParsed", "maxSupportedTransactionVersion": 0}]
    }
    res = requests.post(BASE_URL, json=payload)
    return res.json()["result"]

tx_details = get_transaction_details(signatures[0])
```

```python
KNOWN_DEX_PROGRAMS = {
    "JUP4Fb2cqiRUcaTHdrPC8h2gNsA2ETXiPDD33WcGuJB": "Jupiter",
    "METEOaQo4v6ZqhZ2X4QVfAfH3gfzGm6Bqfb4qaVHQ3v": "Meteora",
    "4ckmDgGz5WqBGkyZvuHMzv3NEpWz5xbgHzrHPG6y9P9": "Raydium",
    "9xQeWvG816bUx9EPta7Xkz2bB1dfP57Zz8kPq8YwTQY": "Serum",
}

def parse_swap_path(tx_data):
    instructions = tx_data['transaction']['message']['instructions']
    swap_path = []
    platforms = []

    for inst in instructions:
        program = inst['programId']
        if program in KNOWN_DEX_PROGRAMS:
            platforms.append(KNOWN_DEX_PROGRAMS[program])
            try:
                swap = inst['parsed']['info']
                source = swap['source']
                destination = swap['destination']
                swap_path.append(f"{source} → {destination}")
            except:
                pass

    return " → ".join(swap_path), list(set(platforms))

path, platforms = parse_swap_path(tx_details)
print("Swap path:", path)
print("DEX platforms:", platforms)
```
